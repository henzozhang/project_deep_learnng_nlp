```python
import time
import random
```

```python
translator = Translator(service_urls=['translate.google.com'], timeout=5)

# Fonction pour traduire une phrase avec une reprise en cas d'erreur
def translate_with_retry(text):
    max_retries = 3  # nombre maximum de réessais
    for i in range(max_retries):
        try:
            result = translator.translate(text, dest='fr')
            return result.text
        except Exception as e:
            print(f"Error while translating '{text}': {e}")
            print(f"Retrying in {i+1} seconds...")
            time.sleep(i+1)  # délai de réessai entre 1 et 3 secondes
    return None

# Appliquer la fonction de traduction sur la colonne 'text' du DataFrame
df['text_fr'] = df['text'].apply(translate_with_retry)

# Afficher le DataFrame avec les traductions
print(df)
```
